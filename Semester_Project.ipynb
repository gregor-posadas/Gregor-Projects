# **Introduction**
Metro Manila is the name given to the Philippines' capial region as well as its largest metropolitan area. The region itself consists of sixteen (16) incredibly urbanized and highly populated cities. It includes Manila, Caloocan, Las Pinas, Makati, Malabon, Mandaluyong, Marikina, Muntinlupa, Navotas, Paranaque, Pasay, Pasig, Quezon City, San Juan, Taguig, and Valenzuela. Collectively, these cities suffer tremendously during extreme weather events. Flooding, in particular, poses the biggest risk to both infrastructure and human life to this region.

According to the Philippine Statistics Authority (PSA), between 100,000 and 120,000 homes were partially damaged due to major disasters in the Philippines. Major disasters include earthquakes, low pressure areas, northeast monsoons, shearlines, and tropical cyclones. In October 2024, 150 people were either dead or missing during the onslaught of Tropical Storm Trami in just a single municipality in the province of Batangas. Imagine the risk to infrastructure and the public if such disasters hit somewhere more urbanized and densely populated.

The challenge brought by such natural disasters - flooding in particular - to both infrastructure and human life calls for an assessment of the country's major population centers as well as an overall risk assessment of these areas when it comes to major hydrological events.

Fortunately, recent advancements in Artificial Intelligence (AI) have provided us with datasets that previously would have taken several years to quantify. Two such datasets that may prove useful in a proposed risk assessment of the Philippines' metropolitan areas are the Waternet dataset and Google Research's Open Buildings dataset. Waternet is a result of a collaboration between Bridges to Prosperity (B2P) and the Better Planet Laboratory. This dataset triples the known extent of mapped waterways globally and was generated by the Waternet AI model. The Open Buildings dataset contains 1.8 billion building detections across 58 million square kilometers in the Global South.

Incorporating both of these datasets into a computational model could help Philippine planners and policymakers understand the vulnerability of the country's major ubran areas to flooding and extreme hydrological events. In addition, incorporating open access datasets such as Giologicx's AEGIS Dataset (a.k.a. Metro Manila Flood Landscape Data) could help give the model more robustness in simulating flood elevations in these urbanized areas.

# **Model Description**

Flooding remains a critical challenge for Metro Manila, where rapid urbanization, inadequate drainage systems, and increasing climate variability converge to amplify the impact of extreme hydrological events. This model aims to integrate state-of-the-art geospatial datasets with computational tools to assess the regionâ€™s resilience to flooding. By combining hydrological data, building information, and flood elevation data, the model provides a comprehensive understanding of the interplay between natural and built environments during flood events.

The model integrates three key datasets:

* **Waternet**: An AI-powered dataset that expands the known extent of global waterways, including previously unmapped streams and rivers. For Metro Manila, Waternet provides a highly detailed mapping of water flow pathways that are critical for understanding hydrodynamic behavior.

* **Open Buildings Dataset**: A high-resolution dataset of building footprints, including their structural characteristics and usage types (e.g., residential, commercial). This enables the model to assess the vulnerability of critical infrastructure to flooding.

* **Metro Manila Flood Landscape Data (AEGIS Dataset)**: This dataset contains flood elevation data for various locations in Metro Manila. It provides crucial information about inundation depths under different flood scenarios, allowing for more accurate modeling of flood impacts on urban areas. This is honestly not the most robust dataset when it comes to flood depths in Metro Manila, though it is derived from raw data from the Philippines' Project NOAH (Nationwide Operational Assessment of Hazards).

The proposed model would consist of the following key processes:

* **Hydrodynamic Simulation**: Using the Waternet dataset, the model simulates water flow dynamics during extreme rainfall events, identifying areas where water accumulates and floods.

* **Flood Elevation Analysis**: The AEGIS Dataset enables the model to incorporate real-world flood elevation data, refining predictions of inundation levels and the spatial extent of flooding.

* **Urban Vulnerability Assessment**: The Open Buildings dataset is used to assess the exposure of built infrastructure to flooding. Vulnerability metrics are calculated based on building density, type, and proximity to water flow paths.

* **Resilience Scoring**: Outputs from the hydrodynamic, elevation, and vulnerability analyses are combined to assign resilience scores to different areas within Metro Manila, highlighting priority zones for intervention.

Traditional flood models like HEC-RAS and SWMM have been widely applied for urban flood studies. While effective, these models often rely on static datasets and lack the granularity required for detailed urban analysis. By incorporating dynamic AI-powered datasets and flood elevation data, this model offers significant advancements in predictive accuracy and spatial resolution.

# **Calibration**

Calibrating our flood risk assessment model is essential to ensure its predictions are accurate and reliable for Metro Manila. Calibration involves fine-tuning key variables and parameters so that the model's outputs align closely with real-world observations.

Below are some key parameters we need to calibrate our
hydrological parameters:

* **Manningâ€™s Roughness Coefficient**: This determines how much resistance water faces as it flows through different types of surfaces. Adjusting this helps simulate realistic water speeds and depths.

* **Inflow Rates**: These are the amounts of water entering the system from rainfall and upstream sources. Calibrating inflow rates ensures our simulations match actual storm events.

* **Water Surface Elevation**: Fine-tuning the starting water levels based on historical flood data helps us predict accurate inundation areas.

* **Runoff Coefficients**: These indicate how much rainfall turns into surface runoff versus soaking into the ground. Adjusting these helps model how quickly areas can flood.

* **Building Resistance Factors**: These factors assess how different types of buildings withstand floodwaters. Calibrating them ensures our vulnerability assessments are spot on.

* **Drainage Capacity**: This determines how effectively the cityâ€™s drainage systems can handle excess water. Adjusting this parameter helps simulate potential drainage failures during extreme events.
